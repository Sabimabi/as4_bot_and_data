### ü•õ **Milk Production: Data Cleaning and Visualization**

- **This notebook loads the dataset `3_–ú–æ–ª–æ–∫–æ.xlsx`, performs data cleaning by removing empty rows and irrelevant entries, and generates a comparison chart to analyze milk production trends.**  
- Fully empty rows are removed  
- The dataset is filtered to include only the relevant columns  
- A grouped bar chart is created to compare 2023 and 2024 milk production by category  


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
import pandas as pd

# üìÇ –ó–∞–≥—Ä—É–∑–∫–∞ Excel-—Ñ–∞–π–ª–∞ —Å –ª–∏—Å—Ç–∞ "7.", –Ω–∞—á–∏–Ω–∞—è —Å 6-–π —Å—Ç—Ä–æ–∫–∏
df = pd.read_excel("data/3_–ú–æ–ª–æ–∫–æ.xlsx", sheet_name="7.", header=5)

# üîç –£–¥–∞–ª—è–µ–º —Å—Ç—Ä–æ–∫–∏, –≥–¥–µ –Ω–µ—Ç –Ω–∞–∑–≤–∞–Ω–∏—è –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ (–ø–µ—Ä–≤—ã–π —Å—Ç–æ–ª–±–µ—Ü –ø—É—Å—Ç–æ–π)
df = df[df[df.columns[0]].notna()]

# ‚úèÔ∏è –ü–µ—Ä–µ–∏–º–µ–Ω–æ–≤—ã–≤–∞–µ–º –ø–µ—Ä–≤—É—é –∫–æ–ª–æ–Ω–∫—É
df = df.rename(columns={df.columns[0]: "–ö–∞—Ç–µ–≥–æ—Ä–∏—è"})

# üì• –û—Å—Ç–∞–≤–ª—è–µ–º —Ç–æ–ª—å–∫–æ 3 –∫–ª—é—á–µ–≤—ã–µ –∫–æ–ª–æ–Ω–∫–∏: –ö–∞—Ç–µ–≥–æ—Ä–∏—è, 2024, 2023
df_filtered = df[["–ö–∞—Ç–µ–≥–æ—Ä–∏—è", df.columns[1], df.columns[2]]].copy()

# üî¢ –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –∑–Ω–∞—á–µ–Ω–∏—è –≤ —á–∏—Å–ª–∞
df_filtered[df.columns[1]] = pd.to_numeric(df_filtered[df.columns[1]], errors='coerce')
df_filtered[df.columns[2]] = pd.to_numeric(df_filtered[df.columns[2]], errors='coerce')

# üßπ –£–¥–∞–ª—è–µ–º —Å—Ç—Ä–æ–∫–∏, –≥–¥–µ –æ–±–µ —è—á–µ–π–∫–∏ (2023 –∏ 2024) –ø—É—Å—Ç—ã–µ
df_filtered = df_filtered.dropna(subset=[df.columns[1], df.columns[2]], how='all')

# üîÑ –°–±—Ä–æ—Å –∏–Ω–¥–µ–∫—Å–æ–≤
df_filtered = df_filtered.reset_index(drop=True)

# üíæ –°–æ—Ö—Ä–∞–Ω—è–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç –≤ –Ω–æ–≤—ã–π —Ñ–∞–π–ª
df_filtered.to_excel("data/3_–ú–æ–ª–æ–∫–æ_cleaned.xlsx", index=False)
print("‚úÖ –ì–æ—Ç–æ–≤–æ! –°–æ—Ö—Ä–∞–Ω–µ–Ω–æ –≤: data/3_–ú–æ–ª–æ–∫–æ_cleaned.xlsx")

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math

# –ó–∞–≥—Ä—É–∑–∫–∞ –¥–∞–Ω–Ω—ã—Ö
df = pd.read_excel("data/3_–ú–æ–ª–æ–∫–æ_cleaned.xlsx")

# –£–∫–æ—Ä–æ—Ç–∏–º –∏ –ø–æ—á–∏—Å—Ç–∏–º –Ω–∞–∑–≤–∞–Ω–∏—è
df["–ö–∞—Ç–µ–≥–æ—Ä–∏—è"] = df["–ö–∞—Ç–µ–≥–æ—Ä–∏—è"].str.strip().str.replace("–∏–∑ –Ω–∏—Ö", "")

# –°–∫–æ–ª—å–∫–æ –∫–∞—Ç–µ–≥–æ—Ä–∏–π –Ω–∞ –∫–∞–∂–¥–æ–º –≥—Ä–∞—Ñ–∏–∫–µ
group_size = 10
num_groups = math.ceil(len(df) / group_size)

# –§—É–Ω–∫—Ü–∏—è –¥–ª—è —Ñ–æ—Ä–º–∞—Ç–∏—Ä–æ–≤–∞–Ω–∏—è –ø–æ–¥–ø–∏—Å–µ–π
def short_number(n):
    if n >= 1_000_000:
        return f"{n/1_000_000:.1f} –º–ª–Ω"
    elif n >= 1_000:
        return f"{n/1_000:.0f} —Ç—ã—Å"
    else:
        return f"{n:.0f}"

# –†–∞–∑–±–∏–≤–∞–µ–º –∏ –≤–∏–∑—É–∞–ª–∏–∑–∏—Ä—É–µ–º
for i in range(num_groups):
    sub_df = df.iloc[i * group_size:(i + 1) * group_size]
    categories = sub_df["–ö–∞—Ç–µ–≥–æ—Ä–∏—è"]
    values_2024 = sub_df[sub_df.columns[1]]
    values_2023 = sub_df[sub_df.columns[2]]
    x = np.arange(len(categories))
    bar_width = 0.35

    plt.figure(figsize=(16, 6))
    bars_2024 = plt.bar(x - bar_width/2, values_2024, bar_width, label="2024 –≥–æ–¥", color="#5DADE2", edgecolor="black")
    bars_2023 = plt.bar(x + bar_width/2, values_2023, bar_width, label="2023 –≥–æ–¥", color="#F5B041", edgecolor="black")

    plt.title(f"ü•õ –ü—Ä–æ–∏–∑–≤–æ–¥—Å—Ç–≤–æ –º–æ–ª–æ–∫–∞: –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ {i * group_size + 1}‚Äì{(i + 1) * group_size}", fontsize=16, weight='bold')
    plt.xlabel("–ö–∞—Ç–µ–≥–æ—Ä–∏—è", fontsize=12)
    plt.ylabel("–û–±—ä—ë–º –º–æ–ª–æ–∫–∞ (–≤ —Ç–æ–Ω–Ω–∞—Ö)", fontsize=12)
    plt.xticks(x, categories, rotation=30, ha="right", fontsize=9)
    plt.yticks(fontsize=10)
    plt.grid(axis='y', linestyle='--', alpha=0.5)
    plt.legend(fontsize=11)

    max_y = max(values_2024.max(), values_2023.max())
    for bars in [bars_2024, bars_2023]:
        for bar in bars:
            height = bar.get_height()
            if not pd.isna(height):
                offset = max_y * 0.01
                pos = height + offset if height > max_y * 0.05 else height + offset * 4
                plt.text(bar.get_x() + bar.get_width()/2, pos, short_number(height),
                         ha='center', va='bottom', fontsize=8, rotation=90)

    plt.tight_layout()
    plt.show()
```
